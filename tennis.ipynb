# Preprocessing

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
pd.options.display.max_columns = 50
tennis = pd.read_csv("data/FrenchOpen-men-2013.csv", index_col = False, na_values="NA")
```

```python
#### Final Points won by player 1 or player 2
Ydata1 = tennis["TPW.1"]
Ydata2 = tennis["TPW.2"]
names = ["FSP", "FSW", "SSP", "SSW", "ACE", "DBF", "WNR", "UFE", "BPC", "BPW", "NPA", "NPW"]
Xdata1 = tennis[["FSP.1", "FSW.1", "SSP.1", "SSW.1", "ACE.1", "DBF.1", "WNR.1", "UFE.1", "BPC.1", "BPW.1", "NPA.1", "NPW.1"]]
Xdata1.columns = names
Xdata2 = tennis[["FSP.2", "FSW.2", "SSP.2", "SSW.2", "ACE.2", "DBF.2", "WNR.2", "UFE.2", "BPC.2", "BPW.2", "NPA.2", "NPW.2"]]
Xdata2.columns = names
Ydata = pd.concat([Ydata1, Ydata2], axis = 0)
Xdata = pd.concat([Xdata1, Xdata2], axis = 0)
Xdata
```

The Tennis Dataset contains information about matches played at the ATP French Open in 2013. The original Dataset consists of 125 observataions on 42 variables. The variables vary from the name of the players, result of the match to common match statistics on both players like first serve percentage. The target variable in our experiments is the number of total points won. To create a meaningful setup specific variables were filtered. Therefore we decided to investigate the impact of player related match statistics like net attempts and first serve percentage on the total points won by the player. This question is reasonable as it deliveres an insight on the reasons for winning points. This leaves us with 250 observations (each observation from the original dataset contains the statistics on both players) on 12 features and the target total points won. There are only four missing values. Outliers only seem to have a meaningful impact on three variables ("ACE", "NPA", "NPW") what represents the different type of plays in tennis.

# Experiments

```python
from experiment_functions import linear_experiment
from sklearn.linear_model import LinearRegression, SGDRegressor, Ridge, Lasso
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import cross_validate
from sklearn.preprocessing import scale, MinMaxScaler
from sklearn import metrics
import random

#### this function can be used for LinearRegression, SGDRegressor, Ridge and Lasso
#    it returns a dictionary with the measurements.mean() given in scoring. classifier, Xdata, Ydata needs to be specified
def linear_experiment(classifier, Xdata, Ydata, scoring = ["explained_variance", "r2" ,"neg_mean_squared_error", "neg_mean_absolute_error"], cv=10):
    #explained variance: how many percent of the variance are explained?
    #r^2: How many percent of the absolut squared error are explained
    random.seed(0)
    results = cross_validate(classifier, Xdata, Ydata, scoring = scoring, cv=cv, return_train_score=False)
    rval = {}
    for i in results:
        rval[i] =  {"mean":results[i].mean(), "std":results[i].std()}
    for i in ["test_neg_mean_squared_error", "test_neg_mean_absolute_error"]:
        if rval[i]["mean"] < 0:
            rval[i]["mean"] = rval[i]["mean"]*-1
    return rval

#### this function can be used for KNeighborsRegressor. It cross_validates KNNRegression by varying n_neigbors form 
#    n_start to n_stop
#    it return a dictionary of lists containing measurement.mean() for each run of n_neighbors
#    classifier does not need to be specified
def knreg_experiment(Xdata, Ydata, scorig=["explained_variance", "r2" ,"neg_mean_squared_error", "neg_mean_absolute_error"],
                     cv=10, n_start = 1, n_stop = 101, weights="uniform", algorithm="brute"):
    random.seed(0)
    ###return object
    rval = {"fit_time_mean":[],
        "fit_time_std":[],
       "score_time_mean":[],
        "score_time_std":[],
       "test_explained_variance_mean":[],
        "test_explained_variance_std":[],
       "test_r2_mean":[],
        "test_r2_std":[],
       "test_neg_mean_squarred_error_mean":[],
        "test_neg_mean_squarred_error_std":[],
       "test_neg_mean_absolute_error_mean":[],
        "test_neg_mean_absolute_error_std":[]
       }
    keys = list(rval.keys())
    ### run cross validate for each n_neighbors
    for i in range(n_start,n_stop):
        results = cross_validate(KNeighborsRegressor(n_neighbors=i, weights = weights, algorithm=algorithm),
                                 Xdata, Ydata, scoring=scoring, return_train_score=False, cv=cv)
        key_index = 0
        for result in results:
            rval[keys[key_index]].append(results[result].mean())
            rval[keys[key_index+1]].append(results[result].std())
            key_index += 2
    ### plot results    
    fig, axlist = plt.subplots(2,3,figsize=(12,8))
    key_index = 0
    for row in range(0, 2):
        for col in range(0,3):
            axlist[row][col].plot(rval[keys[key_index]])
            axlist[row][col].plot(rval[keys[key_index+1]])
            axlist[row][col].set_title(keys[key_index][:-5])
            axlist[row][col].legend(("Mean", "Std"))
            key_index += 2
    return rval
```
