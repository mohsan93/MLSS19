```python
for i in range(0,4):
    tennis.iloc[:,i] = tennis.iloc[:,i].astype("category")
```

```python
#### Final Points won by player 1 or player 2
Ydata1 = tennis["TPW.1"]
Ydata2 = tennis["TPW.2"]
names = ["FSP", "SSP", "SSW", "ACE", "DBF", "WNR", "UFE", "BPW", "NPA", "NPW"]
Xdata1 = tennis[["FSP.1", "SSP.1", "SSW.1", "ACE.1", "DBF.1", "WNR.1", "UFE.1", "BPW.1", "NPA.1", "NPW.1"]]
Xdata1.columns = names
Xdata2 = tennis[["FSP.2", "SSP.2", "SSW.2", "ACE.2", "DBF.2", "WNR.2", "UFE.2", "BPW.2", "NPA.2", "NPW.2"]]
Xdata2.columns = names
Ydata = pd.concat([Ydata1, Ydata2], axis = 0)
Xdata = pd.concat([Xdata1, Xdata2], axis = 0)
Xdata
```

# Statistical Description

```python
fig, axlist = plt.subplots(3,4,figsize=(16,12))
variable = 1
for row in range(0, 3):
    for col in range(0,4):
        axlist[row][col].boxplot(Xdata.iloc[:,variable][Xdata.iloc[:,variable].notnull()])
        axlist[row][col].set_title(Xdata.columns[variable])
        variable += 1
        if variable == Xdata.shape[1]:
            axlist[row][col+1].boxplot(Ydata)
            axlist[row][col+1].set_title("Target: Total Points Won")
            break
```

# Experiments

```python
from experiment_functions import linear_experiment
from sklearn.linear_model import LinearRegression, SGDRegressor
from sklearn.model_selection import cross_validate
from sklearn import metrics
import random


def linear_experiment(classifier, Xdata, Ydata, scoring = ["explained_variance", "r2" ,"neg_mean_squared_error", "neg_mean_absolute_error"], cv=10):
    #explained variance: how many percent of the variance are explained?
    #r^2: How many percent of the absolut squared error are explained
    random.seed(0)
    results = cross_validate(classifier, Xdata, Ydata, scoring = scoring, cv=cv, return_train_score=False)
    rval = {}
    for i in results:
        rval[i] =  results[i].mean()
    for i in ["test_neg_mean_squared_error", "test_neg_mean_absolute_error"]:
        if rval[i] < 0:
            rval[i] = rval[i]*-1
    return rval
```
