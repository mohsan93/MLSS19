```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
#importing the data
train_x = pd.read_csv("X_train.csv")
train_y = pd.read_csv("y_train.csv")
```

## Dataset introdcution

This dataset has been generated by  moveable robot equiped with sensors, it drives over different terrains (concrete, carpets, wood,...)
of which there are a total of 9.
The task here is to predict the surface that the robot is driving on based on the measurement values for linear and angular accelaration,
orientation and so on. It contains 487680 entries and 12 features (of which two are IDs, series_id and measurement_number)

```python
#Training dataset first look
train_x.describe()
```

## Dataset attributes/features

 - **series_id**; Each series represents 128 measurements on a single surface, there are 3809 series in total
 - **measurement_number**; represents the measurement number of the measurement in a certain series, max value is 127, min value 0
 - **orientation_X**; for this task quaternions have been used, this is one of them. They provide a convenient mathematical notation for representing orientations and rotations of objects in three dimensions. Compared to Euler angles they are simpler to compose and avoid the problem of gimbal lock. 
 - **orientation_Y,orientation_Z,orientation_W** are the other three quartions.
 - **angular_velocity_X and Y and Z** represent the angular velocity of the robot
 - **linear_acceleration_X and Y and Z** represent the linear acceleration of the robot
 
 the **train_y** dataset contains the surface of every entry of the train_x dataset, the foreign key is the series_id. It also contains a group_id (can be used for stratified K-Fold, more on that later).

## First look into the data
In the following steps we will look at
 - the number of instances/entries per target surface 
 - the general distribution of the feature values 
 - are there missing values?
 - plot all measurements of a single series to see if we can recognize any patterns
 - correlation plot of the features
 
 

```python
#Checking missing values
train_x.isnull().values.any()
```

```python
#first we need to add the surface (target feature) the the train_x dataframe
surface = []

for entry in train_x['series_id']:
    temp = str(train_y[train_y['series_id'] == entry]['surface'].values)
    surface.append(temp)

#creating a new column called "surface" in  train x containing the surface of each row/entry
train_x['surface'] = surface
```

```python
#function for plotting the feature class distribution
def plot_feature_class_distribution(classes,tt, features):
    plt.figure(figsize=(36, 26))
    for i, col in enumerate(features):
            plt.subplot(4, 3, i+1)
            plt.hist(tt[col], color='blue', bins=100)
            plt.title(col)
    plt.show()

features = train_x.columns.values[3:]
features = features[:-1]
classes = (set(train_y['surface'].values))
plot_feature_class_distribution(classes, train_x, features)
```

As once can see from the plots above, everything is distributed normally except the quaternions. Next, lets plot the distribution of the value distribution of the target feature "surface"

```python
#taking a look at one (time) series:
plt.figure(figsize=(26, 16))
train_x2 = train_x.columns[3:]
for i, col in enumerate(train_x2[:-1]):
    plt.subplot(3, 4, i + 1)
    plt.plot(train_x.loc[train_x['series_id'] == 1, col])
    plt.title(col)
```

No discernable pattern was detected here.

```python
#number of entries/rows per target class value
pd.value_counts(train_y['surface']).plot(kind='barh');
```

We can see that most of the time the robot was driving over concrete, and it rarely drove over hard tiles.

```python
plt.figure(figsize=(25,20))
sns.set(font_scale=1.5)
sns.heatmap(train_x.corr(),annot=True);
```

A positive correlation is just as good as a negative one (absolute correlation has value 1). we can see correlation hotspots between the linear accelearation features (0.38) and between the angualr velocity values.
